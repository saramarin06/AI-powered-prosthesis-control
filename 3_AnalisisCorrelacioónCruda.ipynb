```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfb = df.iloc[:, 0:200]
dfbt=df['Target']
dfb=pd.concat([dfb,dfbt],axis=1)
dfb
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(dfb.corr(), cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```
