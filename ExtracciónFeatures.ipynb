Hago la imputación de datos aplicando interpolación lineal por filas para poder reemplazar los NaN de las Columnas 'Biceps_199' y 'Triceps_199'

```python
# columnas de interés
biceps_cols = [col for col in dataEMG.columns if 'Biceps_' in col]
triceps_cols = [col for col in dataEMG.columns if 'Triceps_' in col]

def interpolate_row(row):
    # copiar fila
    r = row.copy()
    
    # interpolar Biceps
    r[biceps_cols] = r[biceps_cols].interpolate(method='linear', limit_direction='both')
    
    # interpolar Triceps
    r[triceps_cols] = r[triceps_cols].interpolate(method='linear', limit_direction='both')
    
    return r

# aplicar SOLO a filas con NaN
mask_nan = dataEMG[biceps_cols + triceps_cols].isnull().any(axis=1)

dataEMG.loc[mask_nan] = dataEMG.loc[mask_nan].apply(interpolate_row, axis=1)
```

# Extracción de Caracteristicas de Tiempo y Frecuencia 

Se extrae RMS, MAV, VAR, ZC, IMEG, Mean Frequency, Median Frequency y Spectral Entropy de cada ventana EMG.

```python
import numpy as np
from scipy.signal import welch
from scipy.stats import entropy

def extract_emg_features(window, canal, fs=2000):
    """
    Extrae RMS, MAV, VAR, ZC, IMEG, Mean Frequency, Median Frequency,
    y Spectral Entropy de una ventana EMG.
    """

    x = np.array(window, dtype=float)
    N = len(x)
    zc_thresh = 0.001 * np.max(np.abs(window))
    # -------------------------
    # 1) TIEMPO
    # -------------------------

    # RMS
    rms = np.sqrt(np.mean(x**2))

    # MAV
    mav = np.mean(np.abs(x))

    # VAR (Varianza)
    var = np.var(x)

    # Zero Crossing (ZC)
    zc = np.sum(((x[:-1] * x[1:]) < 0) &
                (np.abs(x[:-1] - x[1:]) > zc_thresh))

    # IMEG (Integrated EMG)
    imeg = np.sum(np.abs(x))

    # -------------------------
    # 2) FRECUENCIA
    # -------------------------
    f_psd, psd = welch(x, fs=fs, nperseg=N)

    # Normalizar PSD para entropía
    psd_norm = psd / np.sum(psd)

    # Mean Frequency
    mean_freq = np.sum(f_psd * psd) / np.sum(psd)

    # Median Frequency
    cumulative_power = np.cumsum(psd)
    median_freq = f_psd[np.where(cumulative_power >= cumulative_power[-1] / 2)[0][0]]

    # Spectral Entropy
    spec_entropy = entropy(psd_norm)

    # -------------------------
    # SALIDA
    # -------------------------

    return {
        f"RMS_{canal}": rms,
        f"MAV_{canal}": mav,
        f"VAR_{canal}": var,
        f"ZC_{canal}": zc,
        f"IMEG_{canal}": imeg,
        f"MEANF_{canal}": mean_freq,
        f"MDF_{canal}": median_freq,
        f"SPECTRAL_ENT_{canal}": spec_entropy
    }
```

Automatizo la extracción de carateristicas para todo el dataframe "EMG_Dataset.csv"

```python
df = pd.DataFrame() # Se inicia un dataframe vacio 
for i in range(dataEMG.shape[0]):
    muestra = dataEMG.iloc[i] #iteración por filas (ventanas)
    biceps=muestra[biceps_cols]
    triceps=muestra[triceps_cols]
    target = muestra["Target"]

    features={} # Diccionario donde se guardan la caracteristicas por filas (ventanas)
 #Caracteristicas por musculo y se guardan en el dicicionario 
    features.update(extract_emg_features(biceps,"Biceps"))
    features.update(extract_emg_features(triceps,"Triceps"))
    features["Target"]=target

    EMGfeatures= pd.DataFrame([features]) #Se convierte el diccionario en un Dataframe de una sola fila 

    df = pd.concat([df, EMGfeatures], ignore_index=True) #Sa concatena el dataframe con el df vacio, para obtener el df final 
df
```
