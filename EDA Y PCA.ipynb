```python

df = pd.read_csv("EMG_Features.csv")

print(" Dimensiones:", df.shape)
print("\n Primeras filas:")
display(df.head())

print("\n Información general:")
print(df.info())

print("\n Descripción estadística:")
display(df.describe(include='all'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
numeric_cols = [c for c in numeric_cols if c != "Target"]  # excluir la clase

clases = df["Target"].unique()

plt.figure(figsize=(15, len(numeric_cols) * 3.5))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols), 1, i)
    sns.histplot(data=df, x=col, hue="Target", kde=True, element="step")
    plt.title(f"Distribución de {col} por Clase (Target)")
    plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
corr = df.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap de Correlación")
plt.show()
```

```python
# Correlaciones ordenadas para análisis
corr_pairs = corr.abs().unstack().sort_values(ascending=False)

# Filtrar correlaciones altas (pero no la diagonal = 1.0)
corr_high = corr_pairs[(corr_pairs != 1.0) & (corr_pairs > 0.7)]

print(" Correlaciones fuertes (> 0.7):")
print(corr_high)
```

Luego de realizar la matriz de correlación se puede evidenciar que hay gran correlación entra las caracteristicas: RMS, MAV E IMEG. Por ello se decide realizar PCA para reducir la dimensionalidad y no dar información redundante. 

PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Separar features y target
X = df.drop(columns=["Target"])
y = df["Target"]

# Crear pipeline
pca_pipeline = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("pca", PCA(n_components=0.95))  # asi es para conservar 95% de varianza
])

# Ajustar pipeline a los datos
X_pca = pca_pipeline.fit_transform(X)

print("Shape original:", X.shape)
print("Shape después de PCA:", X_pca.shape)
```

```python
print(" Componentes principales:")
pca = pca_pipeline.named_steps["pca"]
print(pca.components_)
```

```python
pca_etapa = pca_pipeline.named_steps["pca"]
varianza_explicada = pca_etapa.explained_variance_ratio_
print("Varianza explicada por componente:")
print(varianza_explicada)

print("\nVarianza acumulada:")
print(varianza_explicada.cumsum())

print(f"la varianza que se tiene es de {np.sum(varianza_explicada)}")
print("\nNúmero de componentes seleccionados:", pca_etapa.n_components_)   
```

```python
import matplotlib.pyplot as plt

var = pca_etapa.explained_variance_ratio_.cumsum()

plt.figure(figsize=(8,5))
plt.plot(range(1, len(var)+1), var, marker='o')
plt.title("Scree Plot – Varianza Acumulada")
plt.xlabel("Número de Componentes")
plt.ylabel("Varianza Explicada Acumulada")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", palette="viridis")

# Varianza explicada en %
varianza_porcentual = varianza_explicada * 100
componentes = range(1, len(varianza_porcentual)+1)

plt.figure(figsize=(10, 5))
sns.barplot(x=componentes, y=varianza_porcentual, hue=componentes, dodge=False, legend=False)

plt.title("Varianza Explicada por Componente Principal", fontsize=14)
plt.xlabel("Componentes Principales", fontsize=12)
plt.ylabel("Varianza Explicada (%)", fontsize=12)

# Mostrar los valores encima de cada barra
for i, v in enumerate(varianza_porcentual):
    plt.text(i, v + 0.5, f"{v:.1f}%", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
